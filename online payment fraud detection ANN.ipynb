# online payment fraud detection ANN

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data=pd.read_csv("fraud.csv")
```

```python
data
```

```python
data=data.replace(to_replace =["PAYMENT", "CASH_OUT", "PAYMENT", "CASH_IN", "TRANSFER", "DEBIT"],value =[0,1,2,3,4,5])
```

```python
x=data.copy()
x=x.drop("nameOrig", axis=1)
x=x.drop("nameDest", axis=1)
x=x.drop("isFlaggedFraud", axis=1)
x=x.drop("isFraud", axis=1)
x
```

```python
x=x.values
y=data.iloc[:,-2].values
```

```python
#eğitim ve test verisi diye böldük
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
#veri standartlaştırma
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
#yapay sinir ağı modeli kullanacağımızı belirliyoruz
ann=tf.keras.models.Sequential()
#2 gizli katman 1 giriş 1 çıkış katmanı var
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation="relu", input_dim=7))
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#binary = sonucun 0 ya da 1 olmasını sağlıyor. accuracy is doğruluk değerini hesaplar.
# Bir çok optimize yöntemi var. Adam seçildi. optimizerlar Sinapsipler üzerindeki değerlerin nasıl optimize edileceğini belirliyor.
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
#veri setini fit ediyoruz. epochs öğrenme işini yapar. x'ten y'yi öğren
ann.fit(xtrain,ytrain,epochs=10)
```

```python
ypred=ann.predict(xtest)
```

```python
#0.5 üzerinde doğru bildiklerini ata
ypred=(ypred>0.5)
```

```python
#Doğruluk testi dolandırıcılık olmayanlardan 1270867 tanesini doğru, 10 tanesini yanlış bilmiş.
#Dolandırıcılık olanlardna 1001 tanesini doğru 642 tanesini yanlış bilmiş.
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(ytest,ypred)
cm
```

```python
#test verilerini ve çıkan sonuçları karşılaştırıyoruz.
accuracy_score(ytest,ypred)
```
