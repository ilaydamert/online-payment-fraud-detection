# online payment fraud detection ANN

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data=pd.read_csv("fraud.csv")
```

```python
data
```

```python
data=data.replace(to_replace =["PAYMENT", "CASH_OUT", "PAYMENT", "CASH_IN", "TRANSFER", "DEBIT"],value =[0,1,2,3,4,5])
```

```python
x=data.copy()
x=x.drop("nameOrig", axis=1)
x=x.drop("nameDest", axis=1)
x=x.drop("isFlaggedFraud", axis=1)
x=x.drop("isFraud", axis=1)
x
```

```python
x=x.values
y=data.iloc[:,-2].values
```

```python
#eÄŸitim ve test verisi diye bÃ¶ldÃ¼k
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
#veri standartlaÅŸtÄ±rma
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
#yapay sinir aÄŸÄ± modeli kullanacaÄŸÄ±mÄ±zÄ± belirliyoruz
ann=tf.keras.models.Sequential()
#2 gizli katman 1 giriÅŸ 1 Ã§Ä±kÄ±ÅŸ katmanÄ± var
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation="relu", input_dim=7))
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#binary = sonucun 0 ya da 1 olmasÄ±nÄ± saÄŸlÄ±yor. accuracy is doÄŸruluk deÄŸerini hesaplar.
# Bir Ã§ok optimize yÃ¶ntemi var. Adam seÃ§ildi. optimizerlar Sinapsipler Ã¼zerindeki deÄŸerlerin nasÄ±l optimize edileceÄŸini belirliyor.
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
#veri setini fit ediyoruz. epochs Ã¶ÄŸrenme iÅŸini yapar. x'ten y'yi Ã¶ÄŸren
ann.fit(xtrain,ytrain,epochs=10)
```

```python
ypred=ann.predict(xtest)
```

```python
#0.5 Ã¼zerinde doÄŸru bildiklerini ata
ypred=(ypred>0.5)
```

```python
#DoÄŸruluk testi dolandÄ±rÄ±cÄ±lÄ±k olmayanlardan 1270867 tanesini doÄŸru, 10 tanesini yanlÄ±ÅŸ bilmiÅŸ.
#DolandÄ±rÄ±cÄ±lÄ±k olanlardna 1001 tanesini doÄŸru 642 tanesini yanlÄ±ÅŸ bilmiÅŸ.
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(ytest,ypred)
cm
```

```python
#test verilerini ve Ã§Ä±kan sonuÃ§larÄ± karÅŸÄ±laÅŸtÄ±rÄ±yoruz.
accuracy_score(ytest,ypred)
```
